```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
file_path = './data/users_data.csv'
data = pd.read_csv(file_path)

# Clean and preprocess data
data_cleaned = data.copy()

# Remove non-numeric characters from financial columns and convert to numeric
financial_columns = ['per_capita_income', 'yearly_income', 'total_debt']
for col in financial_columns:
    data_cleaned[col] = data_cleaned[col].replace('[\$,]', '', regex=True).astype(float)

# Summary statistics
summary_stats = data_cleaned.describe(include='all').transpose()
print("Summary Statistics:\n", summary_stats)

# Numerical columns for analysis
numerical_columns = ['current_age', 'retirement_age', 'credit_score', 'num_credit_cards'] + financial_columns

# Visualize distributions of numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data_cleaned[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Analyze categorical data
categorical_columns = ['gender']
for col in categorical_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=data_cleaned[col], palette="viridis")
    plt.title(f'Frequency Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
import numpy as np
import math


# Handling Outliers
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Dynamically calculate the grid size for subplots
num_columns = len(numerical_columns)
num_rows = math.ceil(num_columns / 3)  # Assume 3 columns per row

# Visualize Outliers using Boxplots

for i, col in enumerate(numerical_columns, 1):
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data_cleaned[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()    
    plt.savefig(col+".png")
    plt.show()
    
  
```

```python
  

# Detect and print outliers for financial columns
for col in financial_columns:
    outliers = detect_outliers_iqr(data_cleaned, col)
    print(f'Number of outliers in {col}: {len(outliers)}')

# Correlation Analysis
correlation_matrix = data_cleaned[numerical_columns].corr()

# Heatmap of correlations
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Visualizing Relationships
# Example: Yearly Income vs Total Debt
plt.figure(figsize=(15, 10))
sns.scatterplot(data=data_cleaned, x='yearly_income', y='total_debt', hue='credit_score', palette='viridis')
plt.title('Yearly Income vs Total Debt')
plt.xlabel('Yearly Income')
plt.ylabel('Total Debt')
plt.show()


# Example: Credit Score vs Age
plt.figure(figsize=(15, 10))
sns.boxplot(data=data_cleaned, x='current_age', y='credit_score', palette='pastel')
plt.title('Credit Score vs Current Age')
plt.xlabel('Current Age')
plt.ylabel('Credit Score')
plt.show()
```
